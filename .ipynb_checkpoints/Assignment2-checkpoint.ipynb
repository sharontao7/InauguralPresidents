```python
import nltk
import glob
import os
import string
import spacy
from pysyllables import get_syllable_count
import pyphen
import matplotlib.pyplot as plt
import math
```

```python
from collections import defaultdict

path = 'data/'
speeches = dict()
years = dict()

for filename in sorted(glob.glob(os.path.join(path, '*.txt'))):
    filepath = os.path.join(os.getcwd(), filename)
    with open(filepath, 'r') as f:
        year = filename[5:9]
        president_name = filename[10:-4]
        speeches[president_name] = f.read().lower()
        years[president_name] = year
```

## a) Text Length

```python
total_words = dict()
total_sentences = dict()
#sort = sorted(years.items(), key = lambda kv: kv[1])
#sorted_years = OrderedDict()

for president, speech in speeches.items():
    tokens = nltk.word_tokenize(speech)
    sentences = nltk.sent_tokenize(speech)
    #for word in tokens:
        #word = word.replace('-', '')
        
    words = [word for word in tokens if not word.isnumeric()]
    
    total_words[president] = len(words)
    total_sentences[president] = len(sentences)
    
#print(total_words)
#print(total_sentences)
```

```python
from matplotlib.pyplot import figure

plt.figure(figsize=(12, 10))
plt.title("Total Words Per Speech By Year")
plt.xlabel("Year")
plt.ylabel("Words Per Speech")
plt.scatter(list(years.values()), list(total_words.values()))
plt.xticks(rotation = 45)
plt.show()
```

```python
# plotting years vs total_sentences
plt.figure(figsize=(12, 12))
plt.title("Total Sentences Per Speech By Year")
plt.xlabel("Year")
plt.ylabel("Sentences Per Speech")
plt.scatter(list(years.values()), list(total_sentences.values()))
plt.xticks(rotation = 45)
plt.show()
```

## b) Sentence Length

```python
avg = dict()
for president, speech in speeches.items():
    sentences = nltk.sent_tokenize(speech)
    find_avg = sum(len(nltk.word_tokenize(sentence)) for sentence in sentences) / len(sentences)
    avg[president] = find_avg
#print(avg)
```

```python
plt.figure(figsize=(12, 10))
plt.title("Sentence Length Per Speech By Year")
plt.xlabel("Year")
plt.ylabel("Average Words Per Sentence")
plt.scatter(list(years.values()), list(avg.values()))
plt.xticks(rotation = 45)
plt.show()
```

## c) Number of Syllables Per Word

```python
pyphen.language_fallback('nl_NL_variant1')
dic = pyphen.Pyphen(lang='nl_NL')

def avg_num_syllables(text):
    word_tokens = nltk.word_tokenize(text)
    syllables = []
    
    for i in range(len(word_tokens)):
        x = get_syllable_count(word_tokens[i])
        
        if x == None:
            hyphenated = dic.inserted(word_tokens[i])
            hyphen_count = hyphenated.count("-")
            x = hyphen_count+1
        syllables.append(x)

    avg = sum(syllables) / len(syllables)
    return avg


avg_syllables = []
for president, speech in speeches.items():
    avg_syllables.append(avg_num_syllables(speech))
```

```python
plt.figure(figsize=(12, 10))
plt.title("Syllables Per Word Per Speech By Year")
plt.xlabel("Year")
plt.ylabel("Syllables Per Word")
plt.scatter(list(years.values()), list(avg_syllables))
plt.xticks(rotation = 45)
plt.show()
```

## e) Active Vocabulary of the US Presidents:

```python
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
presidents_vocab = dict()

for president, speech in speeches.items():
    count = 0
    active_vocab = []

    doc = nlp(speech)
    for token in doc:
        if token.text.isalpha() and token.text not in stop_words:
            token = token.lemma_
            if token not in active_vocab:
                active_vocab.append(token)

    avg = len(active_vocab) / total_words.get(president)
      
    presidents_vocab[president] = avg
```

```python
plt.figure(figsize=(12, 10))
plt.title("Active Vocabulary Per Speech By Year")
plt.xlabel("Year")
plt.ylabel("Unique Lemmas Per Speech")
plt.scatter(list(years.values()), list(presidents_vocab.values()))
plt.xticks(rotation = 45)
plt.show()
```

## g) Hapax Legomena and Words with a Frequency of at most 4

```python
from collections import Counter
def hapax(words):
  hapax_words = []
  count_words = Counter(words)
  for word in count_words:
    if count_words[word] == 1:
      hapax_words.append(word)
  return hapax_words

def least_freq(words):
  least_freq = []
  count_words = Counter(words)
  for word in count_words:
    if count_words[word] <= 4:
      least_freq.append(word)
  return least_freq

def percent(hap_or_least, words):
  return ((len(hap_or_least) / len(words)) * 100)

hapax_all_speeches = []
hapax_percentage = []

least_freq_all = []
least_freq_percent = []

for president, speech in speeches.items():
    tokens = nltk.word_tokenize(speech)
    hapax_words = hapax(tokens)
    hapax_all_speeches.append(hapax_words)
    
    hapax_percentage.append(percent(hapax_words, tokens))
    
    least_freq__words = least_freq(tokens)
    least_freq_all.append(least_freq__words)
    
    least_freq_percent.append(percent(least_freq__words, tokens))
```

```python
#plot hapax
plt.figure(figsize=(12, 10))
plt.title("Hapax Legomena Per Speech By Year")
plt.xlabel("Year")
plt.ylabel("Hapax Percentage (%)")
plt.scatter(list(years.values()), hapax_percentage)
plt.xticks(rotation = 45)
plt.show()
```

```python
#plot least frequent

plt.figure(figsize=(12, 10))
plt.title("Least Frequent Words Per Speech By Year (<= 4 Frequency Words)")
plt.xlabel("Year")
plt.ylabel("Frequency Percentage (%)")
plt.scatter(list(years.values()), least_freq_percent)
plt.xticks(rotation = 45)
plt.show()
```

## h) Ratio of Percentage Usage of "We" to Percentage Usage of "I"
